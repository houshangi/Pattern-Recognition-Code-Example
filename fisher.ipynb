```python
def fisher_score(sample,label):

    if len(sample) != len(label):
        print('Sample does not match label')
        exit()

    data = pd.concat([sample, label], axis=1)  

    data0 = data[data.diagnosis == 0]
    data1 = data[data.diagnosis == 1]
    n = len(label)
    tag_one_numbers = len(data[data.diagnosis == 1]['diagnosis'].tolist())
    tag_zero_numbers  = len(data[data.diagnosis == 0]['diagnosis'].tolist())
    fisher_score_list = []

    features_list = list(data.columns)[:-1]
    for feature in features_list:

        m0_feature_mean = data0[feature].mean() 

        m0_SW=sum((data0[feature] -m0_feature_mean )**2)

        m1_feature_mean = data1[feature].mean()  
 
        m1_SW=sum((data1[feature] -m1_feature_mean )**2)

        m_all_feature_mean = data[feature].mean()  

        m0_SB = tag_zero_numbers / n * (m0_feature_mean - m_all_feature_mean) ** 2
        m1_SB = tag_one_numbers / n * (m1_feature_mean - m_all_feature_mean) ** 2

        m_SB = m1_SB + m0_SB

        m_SW = (m0_SW + m1_SW) / n
        if m_SW == 0:
            m_fisher_score = np.nan
        else:
            m_fisher_score = m_SB / m_SW

        fisher_score_list.append(m_fisher_score)

    return fisher_score_list
```

```python
score = fisher_score(X,y)
score
```
